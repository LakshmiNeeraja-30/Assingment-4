```python
soldprice_by_age_role=df.groupby(["AGE","PLAYING ROLE"])["SOLD PRICE"].mean().reset_index()
soldprice_by_age_role
```

```python
soldprice_by_age=df.groupby("AGE")["SOLD PRICE"].mean().reset_index()
soldprice_by_age
```

```python
#joining data frames
soldprice_comparision=soldprice_by_age_role.merge(soldprice_by_age,on="AGE",how="outer")
soldprice_comparision
```

```python
#renaming columns
soldprice_comparision.rename(columns={"SOLD PRICE_x":"soldprice_by_age","SOLD PRICE_y":"soldprice_by_age_role"},inplace=True)
```
