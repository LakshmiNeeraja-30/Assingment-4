```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5,10,15,20,25]
plt.scatter(x,y,color="blue",marker="s",s=10,edgecolor="r")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5,10,15,20,25]
color=["red","yellow","blue","black","cyan"]
plt.scatter(x,y,c=color,s=100,edgecolor="green",linewidth=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5,10,15,20,25]
color=["red","yellow","blue","black","cyan"]
size=[70,80,90,100,60]
plt.scatter(x,y,c=color,s=size,marker="s",edgecolor="green",linewidth=2)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5,10,15,20,25]
color=["red","yellow","blue","black","cyan"]
size=[70,80,90,100,60]
plt.scatter(x,y,c=color,s=size,marker="^",edgecolor="green",linewidth=2)
plt.show()
```
