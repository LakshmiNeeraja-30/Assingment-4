```python
import matplotlib.pyplot as plt
Name=["Neerja","kittu","Haseena","Hema","Keerthi"]
Marks=[100,80,90,85,95]
plt.bar(Name,Marks,color="hotpink")
plt.title("Name Vs Marks")
plt.xlabel("Student")
plt.ylabel("Marks")
plt.show()
```

```python
number=[0.1,0.5,1,1.5,2,4,5.5,6,8,9]
plt.hist(number,bins=5,color="red")
plt.xlabel("Number")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
y=np.array([35,50,15])
co=["r","b","hotpink"]
plt.pie(y,colors=co,labels=["DSP","CD","wt"])
plt.legend()
plt.show()
```
